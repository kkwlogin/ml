```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('C:/Users/Student/Downloads/archive (3)/Mall_Customers.csv')
```

```python
df
```

```python
x = df.iloc[:,3:]
x
```

```python
plt.title('Unclustered Data')
sns.scatterplot(x=x['Annual Income (k$)'],y=x['Spending Score (1-100)'])
```

```python
from sklearn.cluster import KMeans, AgglomerativeClustering
```

```python
km = KMeans(n_clusters=4)
```

```python
km.fit_predict(x)
```

```python
#sse
km.inertia_
```

```python
sse =[]
for k in range(1,16):
    km = KMeans(n_clusters=k)
    km.fit_predict(x)
    sse.append(km.inertia_)
```

```python
sse
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming `sse` is a list or array containing your SSE values
sns.lineplot(x=range(1, 16), y=sse)
plt.xlabel('Cluster')
plt.ylabel('SSE')
plt.show()
```

```python
#Method second or alternative for elbow method
from sklearn.metrics import silhouette_score
```

```python
silh =[]
for k in range(2,16):
    km = KMeans(n_clusters=k)
    labels = km.fit_predict(x)
    score = silhouette_score(x, labels)
    silh.append(score)
```

```python
silh
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming `silh` is a list or array with the silhouette values
sns.lineplot(x=range(2, 16), y=silh)
plt.xlabel('Cluster')
plt.ylabel('Silhouette Score')
plt.show()
```

```python
plt.bar(range(2,16,1),silh)
```

```python

km = KMeans(n_clusters=5,random_state=1)
```

```python
labels = km.fit_predict(x)
```

```python
km.labels_
```

```python
cent = km.cluster_centers_
```

```python

import seaborn as sns
import matplotlib.pyplot as plt

plt.title('Clustered Data')

# Scatter plot for the clustered data
sns.scatterplot(x=x['Annual Income (k$)'], y=x['Spending Score (1-100)'], hue=labels)

# Scatter plot for the centroids
sns.scatterplot(x=cent[:, 0], y=cent[:, 1], s=200, color='red')

plt.show()
```
